```python
def get_list_of_equipements(station_id, line_id, direction):
    liste_arret = pd.read_csv("static/arrets-lignes.csv",  sep=";")

    arret = liste_arret[liste_arret["stop_id"] == station_id]

    # Escaliers
    BUCKET = "dlb-hackathon"
    FILE_KEY_S3 = "equipe-2/ratp_localisation_escaliers.csv"
    FILE_PATH_S3 = BUCKET + "/" + FILE_KEY_S3
    with fs.open(FILE_PATH_S3, mode="rb") as file_in:
        escaliers = pd.read_csv(file_in)

    escaliers_par_arret = escaliers.loc[lambda x: x["nom"]
                                        == arret.stop_name.iloc[0]]

    # Ascenseurs
    BUCKET = "dlb-hackathon"
    FILE_KEY_S3 = "equipe-2/etat-des-ascenseurs.parquet"
    FILE_PATH_S3 = BUCKET + "/" + FILE_KEY_S3

    with fs.open(FILE_PATH_S3, mode="rb") as file_in:
        etat_ascenseurs = pd.read_parquet(file_in)

    ascenseurs_par_arret = etat_ascenseurs.loc[lambda x: x["zdcname"]
                                            == arret.stop_name.iloc[0]]
        
    return escaliers_par_arret[["id_escalier", "nom", "type_esc", "localisation"]], ascenseurs_par_arret[["liftid", "zdcname"]]
```

```python
def get_list_of_equipements(zdc):

    # Escaliers
    BUCKET = "dlb-hackathon"
    FILE_KEY_S3 = "equipe-2/ratp_localisation_escaliers.csv"
    FILE_PATH_S3 = BUCKET + "/" + FILE_KEY_S3
    with fs.open(FILE_PATH_S3, mode="rb") as file_in:
        escaliers = pd.read_csv(file_in)

    escaliers_par_arret = escaliers.loc[lambda x:(
        x["nom"] == zdc) & (x["type_esc"] == "EM")]

    # Ascenseurs
    BUCKET = "dlb-hackathon"
    FILE_KEY_S3 = "equipe-2/etat-des-ascenseurs.parquet"
    FILE_PATH_S3 = BUCKET + "/" + FILE_KEY_S3

    with fs.open(FILE_PATH_S3, mode="rb") as file_in:
        etat_ascenseurs = pd.read_parquet(file_in)

    ascenseurs_par_arret = etat_ascenseurs.loc[
        lambda x: x["zdcname"] == zdc
    ]

    return (
        escaliers_par_arret[["id_escalier",
                             "type_esc", "localisation"]],
        ascenseurs_par_arret[["liftid", "liftstatus"]],
    )
```

```python
def get_accessibilite(stop_name):
    BUCKET = "dlb-hackathon"
    FILE_KEY_S3 = "equipe-2/accessibilite-en-gare.parquet"
    FILE_PATH_S3 = BUCKET + "/" + FILE_KEY_S3

    with fs.open(FILE_PATH_S3, mode="rb") as file_in:
        accessibilite = pd.read_parquet(file_in)

    return accessibilite[["stop_name", "accessibility_level_id",
                                   "accessibility_level_name"]].loc[lambda x: x["stop_name"] == stop_name]
```
